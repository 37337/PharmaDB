```python
import pandas as pd
```

```python
drugs_df = pd.read_csv('../DBase/drugs/drugs.tsv', sep='\t')
drugs_df = drugs_df.fillna('')
drugs_df['Generic Names'] = drugs_df['Generic Names'].str.strip()
drugs_df['Generic Names'] = list(map(lambda x: str(x.replace(',"', ', "')), drugs_df['Generic Names']))
```

```python
drugs_df.info()
```

```python
chemicals_df = pd.read_csv('../DBase/chemicals/chemicals.tsv', sep='\t')
chemicals_df = chemicals_df.fillna('')
chemicals_df
```

```python
chemicals_df.info()
```

```python
genes_df = pd.read_csv('../DBase/genes/genes.tsv', sep='\t')
genes_df = genes_df.fillna('')
genes_df['Has CPIC Dosing Guideline'] = list(map(lambda x: str(x.replace('No', '0')), genes_df['Has CPIC Dosing Guideline']))
genes_df['Has CPIC Dosing Guideline'] = list(map(lambda x: str(x.replace('Yes', '1')), genes_df['Has CPIC Dosing Guideline']))
genes_df
```
